# Univariate Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(3, 3))
plt.boxplot(df['salary'])
plt.title('Boxplot of Salary')
plt.show()
```

```python
# as you see that there are many outliers in our model so now we remove them using pandas
```

```python
# Calculate Q1 and Q3
Q1 = df['salary'].quantile(0.25)
Q3 = df['salary'].quantile(0.75)
IQR = Q3 - Q1

# Determine the bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter the DataFrame
df_filtered = df[(df['salary'] >= lower_bound) & (df['salary'] <= upper_bound)]

# Check the shape of the new DataFrame
print(f'Original DataFrame shape: {df.shape}')
print(f'Filtered DataFrame shape: {df_filtered.shape}')
```

```python
# as you see many of outliers are remove from our data
```

```python
plt.figure(figsize=(3, 3))
plt.boxplot(df_filtered['salary'])
plt.title('Boxplot of Salary')
plt.show()
```

```python
plt.figure(figsize=(3, 3))
sns.histplot(df_filtered['salary'])
plt.show()
```

```python
# Create a count plot for Specialization
plt.figure(figsize=(10, 5))
sns.countplot(data=df_filtered, x='specialization', palette='Set2')
plt.xticks(rotation=90)
plt.title('Count of Graduates by Specialization')
plt.xlabel('Specialization')
plt.ylabel('Count of Graduates')
plt.show()
```

```python
# Step 1: Filter for Computer Science Engineering graduates
cs_graduates = df_filtered[df_filtered['specialization'].str.contains("Computer Science", case=False)]

# Step 2: Calculate average salary
average_salary = cs_graduates['salary'].mean()
print(f"Average Salary for Computer Science Engineering graduates: {average_salary}")

# Step 3: Visualize the salary distribution
plt.figure(figsize=(6, 4))
sns.boxplot(data=cs_graduates, y='salary')
plt.title('Salary Distribution for Computer Science Engineering Graduates')
plt.ylabel('Salary')
plt.show()
```

```python
# Step 1: Filter for Computer Science graduates
cs_graduates = df_filtered[df_filtered['specialization'].str.contains("Computer Science", case=False)]

# Step 2: Cross-tabulation
gender_specialization_cs = pd.crosstab(cs_graduates['gender'], cs_graduates['specialization'])

# Step 3: Chi-Square Test
from scipy.stats import chi2_contingency

chi2, p, dof, expected = chi2_contingency(gender_specialization_cs)
print(f"Chi-Square Test p-value for Computer Science Specialization: {p}")

# Step 4: Visualization
plt.figure(figsize=(6, 3))
sns.countplot(data=cs_graduates, x='specialization', hue='gender')
plt.title('Distribution of Computer Science Specializations by Gender')
plt.xticks(rotation=90)
plt.xlabel('Specialization')
plt.ylabel('Count')
plt.legend(title='Gender')
plt.show()
```

```python

from scipy.stats import chi2_contingency

# Test the claim from Times of India regarding the average salary
relevant_roles = ['programming analyst', 'software engineer', 'hardware engineer', 'associate engineer']
relevant_df = df[df['designation'].str.contains('|'.join(relevant_roles), case=False, na=False)]
average_salary_relevant_roles = relevant_df['salary'].mean()
print(f"Average Salary for relevant roles: {average_salary_relevant_roles}")

# Check if there is a relationship between gender and specialization
gender_specialization = pd.crosstab(df['gender'], df['specialization'])
chi2_test_stat, p_value, dof, expected = chi2_contingency(gender_specialization)
print(f"Chi-Square Test p-value for Gender and Specialization: {p_value}")
```

```python
# Conclusion:
# Based on the analysis, we found the following:
# Salary Claim from Times of India: The Times of India article claimed that fresh graduates in roles such as
# Programming Analyst, Software Engineer, Hardware Engineer, and Associate Engineer could earn between ₹2.5-3
# lakhs per annum. Our analysis showed that the average salary for graduates in these roles is approximately
# ₹3.49 lakhs, which is slightly higher than the range stated in the article. This suggests that the claim
# holds, but salaries may be somewhat higher in our dataset.

# Relationship Between Gender and Specialization: The chi-square test results showed a statistically significant
# p-value (p ≈ 0.00000125), indicating a strong relationship between gender and specialization. This implies
# that the choice of specialization is dependent on gender in our dataset. Certain specializations are more
# common among one gender than the other, suggesting a gender-based preference or trend in selecting specializations.
```
